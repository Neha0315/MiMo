```python
zillowdata = pd.read_csv("./HomeHarvest_Past90Days_20241004_183918.csv")
rentals = zillowdata.loc[zillowdata["status"] == "FOR_RENT"]
rentals
```

```python
campus_coords = (41.5043, -81.6084)
rentals_latlon = rentals.dropna(subset = ['latitude', 'longitude'])
def calc_dist(listing_coords, campus_coords):
    return geodesic(listing_coords, campus_coords).miles

rentals_latlon["dist_to_cwru_mi"] = rentals_latlon.apply(lambda row: calc_dist((row['latitude'], row['longitude']), campus_coords), axis=1)
rentals_latlon.head()
```

```python
rentals_edit = rentals_latlon
rentals_edit['list_price'] = rentals_edit['list_price'].fillna((rentals_edit['list_price_min'] + rentals_edit['list_price_max']) / 2)
rentals_edit['price_per_sqft_ann'] = rentals_latlon.apply(lambda row: (row['list_price'] * 12) / row['sqft'], axis=1)
rentals_edit['price_per_sqft_ann'].isna()
```

```python
rentals_final = rentals_edit.dropna(subset = ['sqft', 'price_per_sqft_ann'])
rentals_final.head()
```

```python
rentals_final.columns
```

```python
rentals_final.drop(['list_price_min', 'list_price_max', 'last_sold_date', 'agent_id', 'agent_name', 'agent_email', 'agent_phones', 'agent_mls_set', 'agent_nrds_id', 'broker_id', 'broker_name', 'builder_id', 'builder_name', 'office_id', 'office_mls_set', 'office_name', 'office_email', 'office_phones', 'primary_photo','alt_photos'], axis=1, inplace=True)
rentals_final.head()
```

```python
rentals_final.to_csv("./RentalsCleanedData.csv")
```
